```python
#this is used to change the data type from string to int
def cha_typ (YEARS):
    yea = []
    for i in YEARS:
        yea.append(int(i))
    return yea
```

```python
#this is used to save the changed data type and assign to a variable
new_col = cha_typ(gdp.columns)
```

```python
#this is used to set the new columns
gdp.set_axis(new_col, axis = 1)
```
