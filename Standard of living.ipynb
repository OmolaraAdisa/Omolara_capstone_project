```python
#this is used to change the data type from string to int
def cha_typ (YEARS):
    yea = []
    for i in YEARS:
        yea.append(int(i))
    return yea
```

```python
#visualize the Net migration in Nigeria from 1991-2018
years = data1.columns
net_migration_Nigeria = data1.loc['Nigeria']
plt.xticks(rotation=90)
Nigeria = plt.scatter(years, net_migration_Nigeria, color='green')
plt.xlabel('Date')
plt.ylabel('Value')
```

```python
#visualize the gdp of Nigeria for some years
visual = sns.barplot(gdp.columns, gdp.loc['Nigeria'])
visual.set_title('GDP')
for label in visual.get_xticklabels():
    label.set_rotation(90)
    visual.set(xlabel='Date')
```
